## 1.1 If Statement

```python
# if statement

data = ['A', 'B', 12, 14]

if 'A' in data:
    print("Have data 'A' in list of this (data)")
else:
    print("Doesn't have data 'A' in list of this (data)")
```

```python
# if statement

score = 100
if score >= 80:
    print("Grade A")
elif score >= 70:
    print("Grade B")
elif score >= 60:
    print("Grade C")
elif score >= 70:
    print("Grade D")
else:
    print("You failed!")
```

```python
# Complex statement

score = 100
grade = ('A' if score >= 80 else "B" if score >= 70 else "C" if score >= 60 else "D" if score >= 70 else "failed!")
print(grade)
```

```python
# Email must contain a '.' and '@'
# Email must contain exactly one '@' symbol
# Email must end with '.com', '.org', or '.net'
# Email must not be longer than 254 characters
# Email must start and end with a letter or digit

email = "sutripat@gmail.com"
email = email.strip()
if email =="":
    print("Email cannot be empty.")
# Email must contain a '.' and '@'
elif not('.' in email and '@' in email):
    print("Email must contain . and @")
# Email must contain exactly one '@' symbol
elif email.count('@') != 1:
    print("Email must contain exactly one @.")
# Email must end with '.com', '.org', or '.net'
elif not email.endswith(('.com', '.org', '.net')):
    print("email must end with '.com', '.org', or '.net'")
elif len(email) > 254:
    print("Email must not be longer than 254 characters")
# Email must start and end with a letter or digit
elif not(email[0].isalnum() and email[-1].isalnum()):               # isalnum() => checks if the string or digit has letters
    print("Email must start and end with a letter or digit")
else:
    print("Email is valid.")
```

## 1.2 Match Statement

```python
# match statement: use for equal conditions

# country = "India"
# if country == "United States":
#     print("US")
# elif country == "India":
#     print("IN")
# elif country == "Egypt":
#     print("EG")
# else:
#     print("Unknown country")

country = "United States"
match country:
    case "United States":
        print("US")
    case "India":
        print("IN")
    case "Egypt":
        print("EG")
    case _:
        print("Unknown country")
```

```python
country = "United States"
C_country = ("US" if country=="United States" else "IN" if country=="India" else "EG" if country=="Egypt" else "Unknown country")
print(C_country)
```
