## 1.1 Types

```python
# type (value)
name = 'NEW'
age = 25
print(type(name))
print(type(age))
print(f'my name is {name}, age: {age} year old')

print('=' * 45)
# str()
name = 'NEW'
age = 25
print("I'm " + str(age) + " Year old")
```

## 2. Math

```python
# len(value)
password = '123a'
print(len(password))

if len(password) < 8:
    print("Your password is too short!")
```

```python
# count(substring)
text = """
Python is easy to learn
Python is powerful
Many people love python
"""
print(text.count("python"))
```

## 1.3 Transformations

```python
# replace(old_value, new_value)
price = '1234,56'
price = price.replace(",", ".")
print(price)
```

```python
# string + string
# 'H' + 'i'
first_name = "Sutripat"
last_name = 'Ngamjittrong'
fl_name = first_name + ' ' + last_name
print(f'My name is {fl_name}')
```

```python
# f{}
"""
modern, super-easy way to format and build strings "f"
stands for "formatted" lets you easily put variables and 
expressions directly inside string value
"""

name = "NEW"
age = 25
print(f"My name is {name}, I'm {age} old")
```

```python
# split(pattern) => return to list of character
stamp = "2025-12-10 15:00"
print(stamp.split(" "))
print(stamp.split('-'))
```

```python
# Repetition(*) Operator
# string * <number>
print("ha" * 3)
print('=' * 45)
```

```python
# indexing & slicing
# string[0] => 'cat'[0] / 'cat'[1:3]
texts = "My name is Sutripat, My nickname is NEW"
print(texts[11:19])         # index of 'S' = 11, index of 't' = 18 ( so your code have to 18(+1) )
```

## 1.4 Cleaning

```python
# lstrip() => removes spaces from the left side of a string
# rstrip() => removes spaces from the right side of a string
# strip() => removes spaces of a string
text = " Sutripat NEW "
print(f"My name is{text.lstrip()}.")
print(f"My name is{text.rstrip()}.")
print(f"My name is{text.strip()}.")

print("=" * 45)
text = "###Sutripat NEW###"
print(f"My name is{text.strip("#")}.")
```

```python
# lower() => To change all characters to lowercase
# upper() => To change all characters to uppercase
text = "python PROGRAMMING"
print(text.lower())
print(text.upper())
```

## 1.5 Search

```python
# startwith()
# endwith()
# in => 'a' in 'cat'
phone = '+48-176-1234567'
print(phone.startswith('+49'))      # False
print(phone.endswith('567'))        # True
print('176' in phone)               # True
```

```python
# find()
phone1 = '+48-176-1234567'
print(phone1.find('-'))             # return the first index that find '-'
```

## 1.6 Validation

```python
# isalpha()     => checks if the string has only letters
country = 'USA'
country1 = 'Thailand3'

print(country.isalpha())
print("=" * 45)
print(country1.isalpha())
```

```python
# isnumeric()   => checks if the number(digit) has only letters
phone = "0846592543"
phones = "084-6592543"

print(phone.isnumeric())
print("=" * 45)
print(phones.isnumeric())
```

```python
# isinstance(value, datatype) => checks if a value belonds to a certain datatype
print(isinstance(123, int))
print(isinstance('123', int))
```
