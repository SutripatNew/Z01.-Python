## 1.3) Properties of Line plot

### A) Config color

- 'b' : blue
- 'g' : green
- 'r' : red
- 'c' : cyan
- 'm' : magenta
- 'y' : yellow
- 'k' : black
- 'w' : white

```python
import matplotlib.pyplot as plt
product2 = [15,30,10,20,60]             # ยอดขายสินค้า 2
month = [1,2,3,4,5]                     # เลขเดือน
plt.plot(month, product2, color="m")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

### B.) Config marker

<pre>
- '.'   : point
- 'o'   : cycle
- 'x'   : x marker
- '+'   : plus
- '*'   : star
- 's'   : square
- '|'   : vline
- '-'   : hline
- 'v'   : down
- '^'   : up
- '>'   : right
- '<'   : left
</pre>

```python
import matplotlib.pyplot as plt
product2 = [15,30,10,20,60]             # ยอดขายสินค้า 2
month = [1,2,3,4,5]                     # เลขเดือน
plt.plot(month, product2, color="m", marker='o')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

### C.) Config Line Style

<pre>
- '-'   : solid
- '--'  : dashed
- '-.'  : dash-dot
- ':'   : dotted

</pre>

```python
import matplotlib.pyplot as plt
product2 = [15,30,10,20,60]             # ยอดขายสินค้า 2
month = [1,2,3,4,5]                     # เลขเดือน
plt.plot(month, product2, color="m", marker='o', linestyle=':')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

<pre>
- [marker][line][color]
- [color][marker][line]
</pre>

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4,5], [15,30,10,20,60], 'ro--')     # color marker line
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

### B.) Config Font size Legend

```python
import matplotlib.pyplot as plt

product1 = [10,20,30,40,50]     # ยอดขายสินค้า 1
product2 = [15,30,10,20,60]     # ยอดขายสินค้า 2
month = [1,2,3,4,5]
plt.plot(month, product1, 'ro--')
plt.plot(month, product2, 'bo--')
plt.title('Top Sales 2025', loc='center', color='b', size=20)
plt.xlabel('Month', size=16, backgroundcolor='y')
plt.ylabel('Slaes', size=16, backgroundcolor='lightyellow')
plt.legend(['Mouse', 'Keyboard'], loc=0, fontsize=18)
```

### C.) Edge Color Legend

<pre>
<span style="color:red">facecolor สีของแผ่น legned</span>
- plt.legend([y1,y2], facecolor='red')

<span style="color:red">edgecolor สีของเส้น legend</span>
- plt.legend([y1,y2], edgecolor='red)
</pre>

```python
import matplotlib.pyplot as plt

product1 = [10,20,30,40,50]     # ยอดขายสินค้า 1
product2 = [15,30,10,20,60]     # ยอดขายสินค้า 2
month = [1,2,3,4,5]
plt.plot(month, product1, 'ro--')
plt.plot(month, product2, 'bo--')
plt.title('Top Sales 2025', loc='center', color='b', size=20)
plt.xlabel('Month', size=16, backgroundcolor='y')
plt.ylabel('Slaes', size=16, backgroundcolor='lightyellow')
plt.legend(['Mouse', 'Keyboard'], loc=0, fontsize=18, facecolor='pink', edgecolor='red')
```

### D.) Title Legend

<pre>
- title เป็นการกำหนดข้อความมีค่าเริ่มต้น คือ None

plt.legend([y1,y2], <span style="color:red"> title=..., font_title=..., borderpad=..., labelspacing=0.5</span>)

Remark : borderpad = border size, labelspacing = ระยะห่างระหว่างบรรทัด

</pre>

```python
import matplotlib.pyplot as plt

product1 = [10,20,30,40,50]     # ยอดขายสินค้า 1
product2 = [15,30,10,20,60]     # ยอดขายสินค้า 2
month = [1,2,3,4,5]
plt.plot(month, product1, 'ro--')
plt.plot(month, product2, 'bo--')
plt.title('Top Sales 2025', loc='center', color='b', size=20)
plt.xlabel('Month', size=16, backgroundcolor='y')
plt.ylabel('Slaes', size=16, backgroundcolor='lightyellow')
plt.legend(['Mouse', 'Keyboard'], 
           loc=0, 
           fontsize=12, 
           facecolor='pink', edgecolor='red',
           title='Product', title_fontsize=14,
           borderpad=1.5,
           labelspacing = 2
           )
```
